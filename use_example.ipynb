```python
from simulated_annealing_class import SimulatedAnnealing
import numpy as np
from matplotlib import pyplot as plt 
```

```python
# Função que será maximizada
def function(x):
    if x < -1 or x > 1:
        y = 0
    else:
        y = np.cos(50*x) + np.sin(20*x)

    return y
```

```python
# Função de avaliação que retorna o valor negativo da função function(x)
def evaluation_function(x):
    return -function(x)
```

```python

# Função de vizinhança que gera uma nova solução adicionando um valor aleatório à solução atual
def neighborhood_function(x, scale=0.1):
    return x + np.random.uniform(-scale, scale)
```

```python
# Parâmetros do algoritmo Simulated Annealing
initial_temperature = 5
cooling_rate = 0.99
max_iterations = 5000
final_temperature = 0.001
initial_solution = np.random.uniform(-1, 1)
```

```python
# Criação do objeto SimulatedAnnealing
sa = SimulatedAnnealing(evaluation_function, neighborhood_function, initial_temperature, cooling_rate, max_iterations, final_temperature, initial_solution)
```

```python
# Execução do algoritmo e obtenção da solução encontrada
best_solution = sa.run()

# Exibição do ponto de máximo encontrado
print("Ponto de máximo encontrado: x =", best_solution, " y =", function(best_solution))
```

```python
# Mostrando o valor encontrado no gráfico
X = np.linspace(-1 , 2, num=1000)
function = np.vectorize(function)
plt.plot(X, function(X))
plt.scatter(best_solution,function(best_solution), marker='x')
```
